## Zipf's law demo from the command line

This pipeline:
1. lowercases the text
2. tokenizes into one word per line
3. counts word frequencies
4. sorts by frequency
5. plots rank vs. frequency in the terminal

```python
%%bash
cat ulysses.txt | tr 'A-Z' 'a-z' | tr -sc 'a-z' '\n' | sort | uniq -c | sort -nr | \
python -c "
import sys, plotext as plt
freq = [int(line.split()[0]) for line in sys.stdin if line.strip()]
rank = range(1, len(freq)+1)
plt.scatter(rank, freq)
plt.xscale('log')
plt.yscale('log')
plt.title('Zipf Law (Ulysses)')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.show()
"
```
